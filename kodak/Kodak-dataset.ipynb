```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

def load_image_pairs(input_dir, gt_dir):
    """
    Load corresponding images from input and groundtruth directories.
    Returns two lists of loaded images.
    """
    # Get list of files in input directory
    input_files = sorted([f for f in os.listdir(input_dir) if f.endswith('.png')])
    
    input_images = []
    gt_images = []
    
    for filename in input_files:
        # Load input image
        input_path = os.path.join(input_dir, filename)
        input_img = Image.open(input_path)
        input_images.append(np.array(input_img))
        
        # Load corresponding groundtruth image
        gt_path = os.path.join(gt_dir, filename)
        gt_img = Image.open(gt_path)
        gt_images.append(np.array(gt_img))
        
    return input_images, gt_images

def visualize_pair(input_img, gt_img, index):
    """
    Display an input image and its corresponding groundtruth image side by side.
    """
    plt.figure(figsize=(12, 6))
    
    # Display input image
    plt.subplot(1, 2, 1)
    plt.imshow(input_img)
    plt.title(f'Moisaiced Image')
    plt.axis('off')
    
    # Display groundtruth image
    plt.subplot(1, 2, 2)
    plt.imshow(gt_img)
    plt.title(f'Ground Truth Image')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

```

```python
# Set up paths
input_dir = '../dataset/kodak/input'
gt_dir = '../dataset/kodak/groundtruth' 
# Load all image pairs
input_images, gt_images = load_image_pairs(input_dir, gt_dir)

# Visualize the first two pairs
for i in range(2):
    visualize_pair(input_images[i], gt_images[i], i+1)

print(f"Total number of image pairs loaded: {len(input_images)}")
```
